## 2. Variables (Cajas de memoria)

Una variable es como una **caja** donde guardamos un dato y le ponemos una etiqueta (nombre) para encontrarlo después.

Existen varios tipos de datos básicos:
* **String (Texto):** Va entre comillas `"Hola"`.
* **Int (Entero):** Números sin coma `10`.
* **Float (Flotante):** Números con coma `10.5`.

```python
# Creando variables
nombre_alumno = "Juan"
edad = 25
altura = 1.75
es_estudiante = True  # Esto es un Booleano (Verdadero/Falso)
print("Variables creadas.")
print(type(nombre_alumno))  # Tipo de dato: str (cadena de texto)
print(type(edad))           # Tipo de dato: int (e"¿Cuánntero)
print(type(altura))        # Tipo de dato: float (número decimal)           
print(type(es_estudiante))  # Tipo de dato: bool (booleano)     

# Usando las variables

print("Nombre:", nombre_alumno)
print("Edad:", edad)
```

## 3. La Entrada: `input()`

Hasta ahora, nosotros escribimos los datos en el código. Pero, ¿y si queremos preguntarle algo al usuario?

Usamos `input()`. **OJO:** `input` siempre devuelve texto (string). Si queremos números, hay que convertirlo con `int()`.

```python
# Preguntando nombre
tu_nombre = input("¿Cómo te llamas? ")
print("¡Hola " + tu_nombre + "! Gusto en conocerte.")
```

```python
# Preguntando un número (Conversión necesaria)
edad_texto = input("¿Cuántos añ"¿Cuános tienes? ")
edad_numero = int(edad_texto)  # Convertimos texto a número

# Ahora podemos hacer matemáticas
edad_futura = edad_numero + 10
print("En 10 años tendrás:", edad_futura)
```

## 4. Lógica y Decisiones: `if / else`

Los programas no son lineales. A veces tienen que tomar caminos diferentes dependiendo de los datos. Para eso usamos:

* `if` (Si pasa esto...)
* `else` (Si no...)
* `elif` (Sino, si pasa esto otro...)

**Importante:** ¡La identación (sangría/espacio) es vital en Python!

```python
temperatura = 35

if temperatura > 30:
    print("🔥 ¡Hace mucho calor! Prende el ventilador.")
elif temperatura < 10:
    print("❄️ ¡Hace frío! Ponte abrigo.")
else:
    print("☁️ El clima está agradable.")
```

## 🏆 Ejercicio Práctico: Calculadora de Edad

**Consigna:**
1. Crea un programa que le pregunte al usuario su **año de nacimiento**.
2. Calcula automáticamente cuántos años tiene (basado en el año actual, por ejemplo, 2025).
3. Usa un condicional `if`:
    * Si es mayor o igual a 18 años, imprime: **"Eres mayor de edad, puedes pasar."**
    * Si es menor de 18 años, imprime: **"Eres menor, acceso denegado."**

```python
# Escribe tu código aquí debajo 👇

# 1. Definir año actual
anio_actual = 2026

# 2. Pedir año de nacimiento (recuerda convertir a int)
aniodenac = int (input("¿En qué año naciste?"))

# 3. Calcular edad
edad = anio_actual - aniodenac

# 4. Condicional If / Else
if edad >= 18:
    print("Puedes pasar, eres mayor de edad")
else:
    print("No puedes pasar, eres menor de edad")

```
